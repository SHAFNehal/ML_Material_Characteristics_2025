```python
# Import required library
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Setting the style for the plots
sns.set_theme(style="darkgrid")
```

```python
# Read data
data = pd.read_excel('./data/Formatted_Data_v2.xlsx')
```

```python
# Boxplots for categorical variables against Sample_Shrinkage
categorical_variables = ['Sample_Size', 'Sample_Prints', 'Sample_Geometry', 'Sample_Delay']

fig, ax = plt.subplots(1, len(categorical_variables), figsize=(15, 5))

for i, var in enumerate(categorical_variables):
    sns.boxplot(x=var, y='Sample_Shrinkage', data=data, ax=ax[i])
    ax[i].set_title(f'Sample Shrinkage vs {var}')
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=45)


plt.tight_layout()
plt.show()
```

```python
# Boxplots for categorical variables against Sample_Shrinkage
categorical_variables = ['Sample_Size', 'Sample_Prints', 'Sample_Geometry']

fig, ax = plt.subplots(1, len(categorical_variables), figsize=(15, 5))

for i, var in enumerate(categorical_variables):
    sns.boxplot(x=var, y='Sample_Relative_Density', data=data, ax=ax[i])
    ax[i].set_title(f'Sample_Relative_Density vs {var}')
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
